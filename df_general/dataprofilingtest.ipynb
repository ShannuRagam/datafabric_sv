```python
import pandas as pd
import json
from great_expectations.dataset import PandasDataset
from great_expectations.profile.basic_dataset_profiler import BasicDatasetProfiler

df = pd.read_csv("bank_transaction.csv")
dataset = PandasDataset(df)
profiler = BasicDatasetProfiler()
expectation_suite, validation_result = profiler.profile(dataset, run_id="profiling_run")
```

```python
validation_result
```

```python
lst=[]
for i in validation_result["results"]:
    if i["expectation_config"]["expectation_type"] not in lst:
        lst.append(i["expectation_config"]["expectation_type"])
print("\n".join(lst))
        
```
