```python
import findspark
findspark.init()

import pyspark # only run after findspark.init()
from pyspark.sql import SparkSession
spark = SparkSession.builder.getOrCreate()

df = spark.sql('''select 'spark' as hello ''')
df.show()
```

```python
dfn= spark.read.csv(r'C:\Users\venkatesh.ragam\Downloads\amazon_sales.csv',header=True,inferSchema=True)
dfn.count()
```

```python
from pyspark.sql.functions import *
from pyspark.sql.types import *
df2 = dfn.withColumn('discounted_price',regexp_replace(col('discounted_price'), ",", "").cast(DoubleType()))
df2.schema
```

```python
df2.write.mode('overwrite').option("header",True).csv(r'C:\Users\venkatesh.ragam\Downloads\as')
```

```python
col_df= df2.select('discounted_price')
rdd = col_df.rdd
print(rdd.collect())
```

```python
rows_with_type = dfn.rdd.map(lambda row: (row.discounted_price, type(row.discounted_price).__name__)).collect()

for row, data_type in rows_with_type:
    print(f"Row {row} data type: {data_type}")
```
